# Openning Data

```julia
using SparseArrays
using JuMP
using Ipopt
using GLPK
using LinearAlgebra
using MAT
```

```julia
file = matopen("R4T3.mat")
varnames = names(file)
```

```julia
S = read(file, "S")
L = read(file, "L")
U = read(file, "U")
lambda = read(file,"lambda")
```

```julia
(m,n)=size(S)
(r,c) = size(L)
```

# constructing model

```julia
mzeros = zeros(m,1);
```

```julia
function opt_vector(S,l,u,z)
    model = Model(Ipopt.Optimizer)
    @variable(model, l[i] <= v[i = 1:r] <= u[i])
    @variable(model,a[1:r])
    @constraint(model, S*v .== mzeros)
    @constraint(model, a .>= v)
    @constraint(model, a .>= -v)
    @NLobjective(model,Min,sum(sqrt(1+z[i]+a[i]) for i=1:r))
    optimize!(model)
    
    return JuMP.value.(v)
end
```

```julia
function zero_rows(V)
    indexes= zeros(Int64,size(V)[1],1)
    k=1
    for i in 1:size(V)[1]
        if sum(V[i,:].*V[i,:])<1e-10
            indexes[k]= i
            k+=1
        end
    end
    return length(indexes[1:k-1])
end
```

```julia
function row_squares(V)
    r,c = size(V,1), size(V,2)
    squares = zeros(r,1)
    for i in 1:r
        z=0
        for j in 1:c
           z += (V[i,j])^2
        end
        #squares[i]=sqrt(z)
        squares[i]=z
    end
    return squares
end
```

```julia
l = L[:,1]
u = U[:,1]
z = ones(r,1)
v = opt_vector(S,l,u,z)
zero_rows(v)
```

```julia
for i in 1:c
    if isdefined(Main, :IJulia)
        Main.IJulia.stdio_bytes[] = 0
    end
    l = L[:,i]
    u = U[:,i]
    v = opt_vector(S,l,u,z)
    z = z .+ (v.*v)
    V = [V v]
    print("**i : ",i,"\n")
    print("***",zero_rows(V),"\n") 
end
```
